## TCP/IP 클라이언트 예제

사용자가 입력한 메세지를 서버로 전송하고, 서버의 응답을 받아 출력하는 간단한 클라이언트 입니다.

```python
import socket
```

```python
# 1. 서버 접속 정보 설정

HOST = '192.168.0.49' # 서버의 IP 주소 (localhost)
PORT = 9888 # (0~ 65535 중 하나, 다른 서비스와 중복 금지) 

# 2. 소켓 생성

client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM) # AF_INET 는 소켓이 통신할 수 있는 주소 유형을 지정 --> IPv4 주소

# 3. 서버에 연결 시도

client_socket.connect((HOST,PORT))
print(f"서버 {HOST}:{PORT}에 연결되었습니다.")
print("메세지를 입력하세요. 종료하시려면 'exit'를 입력\n")


# 4. 메세지 송수신 루프
while True:
    message = input("보낼 메세지 : ") # 사용자 입력 대기
    
    # exit 입력시 종료
    if message.lower() == 'exit':
        client_socket.sendall(message.encode()) # 서버에 종료 알림
        break
    
    client_socket.sendall(message.encode())

    # 서버로부터 응답 수신
    data = client_socket.recv(1024).decode() #최대 1024 byte씩. byte를 str으로 변환해서 읽어옴
    print(f"서버 응답 : {data} \n")

# 5. 소켓 종료
client_socket.close()
print("클라이언트 종료 완료")
```
