클라이언트의 접속을 기다리며, 클라이언트가 보낸 메세지를 수신하고 간단한 응답을 보내는 서버입니다.

```python
import socket # 네트워크 통신을 위한 기본 모듈
```

```python
# 1. 서버 기본 설정

HOST = '192.168.0.49'
PORT = 9888 # (0~ 65535 중 하나 선택) 


# 2. 소켓 생성

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# 3. IP 와 포트를 소켓에 바인딩(연결)

server_socket.bind((HOST,PORT)) # server 가 클라이언트 요청을 받을 수 있도록 설정
#클라이언트에서도 서버의 호스트와 포트 정보를 가지고 접속할 수 있음

# 4. 클라이언트 연결 대기 시작

server_socket.listen()
print(f"서버가 {HOST}:{PORT}에서 연결 대기 중입니다..")

#5. 클라이언트 연결 수락 (accept)

client_socket, addr = server_socket.accept()
print(f"클라이언트 {addr} 연결 완료")

# 6. 클라이언트와 메세지 송수신 루프
# 1회성으로 송/수신이 아니라 계속 송수신 할 수 있도록 루프 만들어주기

while True:
    data = client_socket.recv(1024).decode()
    if not data:
        print("데이터 수진 송료 (클라이언트 연결 해제됨)") # recv(1024) 최대 받을 수 있는 # decode() byte-> str
        break

    if data.lower() == 'exit':
        print("클라이언트 종료 요청 수신")
        break

    #수신된 메세지 출력
    print(f"클라이언트 메세지 : {data}")

    #서버의 응답 생성
    reply = f"서버 응답 : [{data}] 잘 받았습니다."
    client_socket.sendall(reply.encode())

    # 클라이언트로 응답 전송 (str -> bytes 로 변환해서 보내줌)

# 7. 연결 종료

client_socket.close()
server_socket.close()
print("서버 종료 완료")
```
